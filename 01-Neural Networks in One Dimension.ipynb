```python
import torch
import torch.nn as nn
from torch import sigmoid
```

```python
class Net(nn.Module):
    def __init__(self, D_in, H, D_out):
        super(Net, self).__init__()
        self.linear1 = nn.Linear(D_in, H)
        self.linear2 = nn.Linear(H, D_out)
    def forward(self, x):
        x = sigmoid(self.linear1(x))
        x = sigmoid(self.linear2(x))
        return x
```

```python
#D_in = 1, H = 2, D_out = 1
model = Net(1,2,1)
```

```python
x = torch.tensor([0.0])
yhat = model(x)
yhat
```

```python
X = torch.tensor([[0.0], [2.0], [3.0]])
yhat = model(X)
yhat
```

***Train the model***

```python
X = torch.arange(-20, 20, 1).view(-1,1).type(torch.FloatTensor)
Y = torch.zeros(X.shape[0])
Y[(X[:, 0] > 4) & (X[:,0] < 4)] = 1.0
```

```python
X.shape, Y.shape
```

```python
Y = Y.unsqueeze(1)
Y.shape
#https://stackoverflow.com/questions/57798033/valueerror-target-size-torch-size16-must-be-the-same-as-input-size-torch
```

```python
def train(Y, X, model, optimizer, criterion, epochs = 1000):
    cost = []
    total = 0
    for epoch in range(epochs):
        total = 0
        for y, x in zip(Y,X):
            yhat = model(x)
            loss = criterion(yhat, y)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            #cumulative loss
            total += loss.item()
        cost.append(total)
    return cost
```

```python
criterion = nn.BCELoss()
```

```python
model = Net(1,2,1)
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
cost = train(Y, X, model, optimizer, criterion, epochs = 1000)
```
